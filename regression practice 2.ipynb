```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
def model (x, a, b, c):
    return a*x**2+b*x+c
t= [0, 0.04, 0.08, .12, .16, .2, .24]
y= [0, 3.3, 5.8, 7.7, 9.1, 9.9, 10.2]
xx= np.linspace(0.,0.45,20)
 
a,b,c = curve_fit(model, t, y)[0]
print (a,b,c)
plt.plot(xx, model(xx,a, b, c), label='y=-182.291x^2+85.804x+0.06')

plt.legend (loc=3)
plt.plot (t,y,'ro')
plt.ylabel ('cm')
plt.xlabel ('seconds')
plt.title ('Basketball Bouncing')
```
